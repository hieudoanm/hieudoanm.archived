```python
from bs4 import BeautifulSoup
import pandas
import requests
from tqdm.notebook import tqdm as tdqm_notebook
```

## Countries

```python
COUNTRIES_URL = "https://restcountries.com/v3/all"
response = requests.get(COUNTRIES_URL)
countries = response.json()
```

```python
countries_data_frame = pandas.DataFrame(countries)
countries_data_frame.to_csv("./countries/all.csv", index=False, header=True)
countries_data_frame
```

## United Nation Members

```python
united_nation_countries_data_frame = countries_data_frame[countries_data_frame["unMember"]]
united_nation_countries_data_frame
```
