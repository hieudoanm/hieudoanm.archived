```python
import pandas
import requests
```

```python
URL = "https://api.coinranking.com/v2/coins"
```

```python
coins = []

for i in range(0, 10):
  offset = i * 100
  response = requests.get(f"{URL}?limit=100&offset={offset}")
  print(offset)
  if response.status_code == 200:
    data = response.json()
    coins_per_page = data.get("data", {}).get("coins", [])
    coins = coins + coins_per_page

len(coins)
```

```python
coins_data_frame = pandas.DataFrame(coins)
coins_data_frame.to_csv("./coins.csv", index=False)
coins_data_frame
```
